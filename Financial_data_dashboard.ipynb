```python
import pandas as pd
import yfinance as yf
import streamlit as st
import plotly
import plotly.graph_objects as go
```

```python
ticker = 'AAPL' #Apple ticker
data = yf.download(ticker, start='2018-01-01', end='2025-01-01',group_by='column')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
#Adding moving average to reduce noise
data['MA20'] = data['Close'].rolling(window=20).mean()
# Filtering
filtered = data.loc['2022-01-01':'2024-12-31']
filtered.head()
```

```python
filtered.shape
```

```python
print(type(filtered.index))
```

```python
filtered.columns = [col[0] if col[1] == '' else f"{col[0]}" for col in filtered.columns]
print(filtered.columns)
```

```python
import plotly.graph_objs as go


fig = go.Figure()
fig.add_trace(go.Scatter(
    x=filtered.index,
    y=filtered['Close'],
    name='Close',
    mode='lines'
))
fig.add_trace(go.Scatter(
    x=filtered.index,
    y=filtered['MA20'],
    name='MA20',
    mode='lines'
))
fig.update_layout(
    title=f'{ticker} Price & MA20',
    xaxis_title='Date',
    yaxis_title='Price',
    template='plotly_white',
    legend=dict(x=0, y=1)
)
fig.show()
```
